```python
import json
import urllib.request as req
import pandas as pd

url = "https://api.gamer.com.tw/anime/v1/danmu.php?videoSn=48443&geo=TW%2CHK"
f = req.urlopen(url)
content = f.read()
content = json.loads(content)
danmu = content["data"]["danmu"]
df = pd.DataFrame(danmu)
df.to_csv("danmu.csv", encoding="utf-8")
df
```

```python
import pandas as pd
import os
# 要有資料夾才能存檔,須先創建
dn = "baha"
#如果資料夾不存在 就開起來
if not os.path.exists(dn):
    os.makedirs(dn)
# 加這個索引不會變成多一欄 index_col=0
df = pd.read_csv("danmu.csv", encoding="utf-8", index_col=0)

for u in df["userid"].unique():
    fil = df["userid"] == u
    fn = dn + "/" + u + ".csv"
    df[fil].to_csv(fn, encoding="utf-8")
```
