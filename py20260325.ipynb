```python
scores = [80, 60, 70, 50]

count = 0

i = 0
while i < len(scores):
    s = scores[i]
    if s >= 60:
        count = count + 1
    i = i + 1
print(count)
```

```python
scores = [80, 60, 70, 50]

count = 0

for s in scores:
    if s >= 60:
        count = count + 1
    i = i + 1
print(count)
```

```python
# 1 加到 10

total = 0
i = 0
while i < 10:
    total = total + (i + 1)
    print(total)
    i = i + 1
#print(total)
```

```python
# 1 加到 10
# for in 後面接 list[] 或 string 或 range()
# range(3)  ->0,1,2
# range(2,5) ->2,3,4
# range(1,7,2)  ->1,3,5

total = 0
i = 0
for i in range(10):
    total = total + (i + 1)
    print(total)
```
