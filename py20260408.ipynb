```python
# pandas 函式庫 專門處理表格
import json
import urllib.request as req

url = "https://api.gamer.com.tw/anime/v1/danmu.php?videoSn=48213&geo=TW%2CHK"
f = req.urlopen(url)
content = f.read()
# 型態轉換: bytes => list/dict(JSON)
#print(type(content))
#loads的s表示字串(動詞後面的s通常表示str)
content = json.loads(content)
danmu = content["data"]["danmu"]
print(danmu)

for d in danmu:
    d_userid = d["userid"]
    d_text = d["text"]
    print(d_userid, d_text)
```

```python
# pandas 函式庫 專門處理表格
import json
import urllib.request as req
import pandas as pd

url = "https://api.gamer.com.tw/anime/v1/danmu.php?videoSn=48213&geo=TW%2CHK"
f = req.urlopen(url)
content = f.read()

content = json.loads(content)
danmu = content["data"]["danmu"]
df = pd.DataFrame(danmu)
df.to_csv("danmu.csv", encoding="utf-8")
```
