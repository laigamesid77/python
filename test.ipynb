```python
import random

all = ["棒子","老虎","雞","蟲"]
my = int(input("0,1,2,3:"))
com = random.randint(0,3)

print("我出:"+all[my])
print("電腦出:"+all[com])

if com == my + 1 % 4:
    print("WIN")
elif my == com + 1 % 4:
    print("LOSE")
else :
    print("EVEN")
```
