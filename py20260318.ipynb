```python
import random
# 兩組猜拳遊戲 同一組對決,不同組為平手
trans = ["剪刀[0]", "石頭[0]", "布[0]", "剪刀[1]", "石頭[1]", "布[1]"]
#com = random.randint(0, 5)
com = 0
my = int(input("0.剪刀[0], 1.石頭[0], 2.布[0], 3.剪刀[1], 4.石頭[1], 5.布[1]："))
print("我:"+trans[my])
print("電腦:"+trans[com])

# 設定商相同為同組
if my // 3 == com // 3:
    #將第2組改為第1組模式
    my = my % 3
    com = com % 3
    if my == (com + 1) % 3:
        print("WIN")
    elif com == (my + 1) % 3:
        print("LOSE")
    else :
        print("EVEN")
else:
    print("你們不同組平手")
```
