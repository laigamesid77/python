```python
import ColabTurtle.Turtle as turtle
turtle.initializeTurtle(initial_speed=10, initial_window_size=(800,500))


# 設定幾邊形
n = 50
# 設定往前走多少
m = 60
# 設定角度
angle = 360 / n

for i in range(n):
    # 拿到位置: turtle.pos()
    #取得起始位置
    x, y = turtle.pos()
    #往前走
    turtle.forward(m)
    #取得第二個位置
    oldx, oldy = turtle.pos()
    #回到起始位置
    turtle.goto(x, y)
    #右轉
    turtle.right(angle)
    #往前走
    turtle.forward(m)
    #前往第二個位置
    turtle.goto(oldx, oldy)
    #前往起始位置
    turtle.goto(x, y)

```

```python
import random
import ColabTurtle.Turtle as turtle
turtle.initializeTurtle(initial_speed=10, initial_window_size=(800,500))

n = 100
angle = 360 / n
m = 100

i = 0
while i < n:
    # 等號（=）只會執行一次,要多次要寫在迴圈裡
    r = random.randint(0, 255)
    g = random.randint(0, 255)
    b = random.randint(0, 255)
    turtle.color(r, g, b)
    #

    x, y = turtle.pos()
    turtle.forward(m)
    x1, y1 = turtle.pos()
    turtle.goto(x, y)
    turtle.right(angle)
    turtle.forward(m)
    turtle.goto(x1, y1)
    turtle.goto(x, y)
    i = i + 1
```

```python
import random

trans = ["L", "E", "W"]
win, even, lose = 0, 0, 0
i = 0
while i < 10:
    # -1: lose 0: even 1: win
    result = random.randint(-1, 1)
    #因為要把-1,0,1改成0,1,2 所以要+1
    print(trans[result+1])
    if result == -1:
        lose = lose + 1
    elif result == 0:
        even = even + 1
    elif result == 1:
        win = win + 1
    i = i + 1
print(lose, even, win)
```

```python
# 不固定次數的迴圈(其他條件)
# while
import random

trans = ["L", "E", "W"]
win, even, lose = 0, 0, 0

# 先把遊戲簡化: 只有我贏三次才結束(當我贏的次數小於三次就要繼續玩)
# while win < 3
# 組合不同條件
# 布林(是否)型態: True False
# 1. 否: not 5 == 3
# 2. 組合: and(而且) or(或者)
# 我贏三次 或者 對方贏三次就結束
# 我還沒贏三次 而且 對手也還沒贏三次 就要繼續玩
while True: #當條件太多時,無腦用 while True ,在用break跳出(可以將判斷寫在下面比較方便)
    # -1: lose 0: even 1: win
    if win == 3:
        print("win")
        break
    elif lose == 3:
        print("lose")
        break
    result = random.randint(-1, 1)
    print(trans[result+1])
    if result == -1:
        lose = lose + 1
    elif result == 0:
        even = even + 1
    elif result == 1:
        win = win + 1

print(lose, even, win)
```

```python
import random

trans = ["L", "W"]
win, lose = 0, 0
# 贏2分以上才算贏
# abs()
score_win = 5
while True:
    result = random.randint(0, 1)
    print(trans[result])
    if result == 1:
        win = win + 1
    elif lose == 0:
        lose = lose + 1
    #如果我贏達到5分而且贏對手2分
    if win >= score_win and win >= (lose + 2):
        print("win")
        break
print(win, lose)
```
