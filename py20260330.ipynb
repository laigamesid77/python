```python
import ColabTurtle.Turtle as turtle
turtle.initializeTurtle(initial_speed=10, initial_window_size=(800,500))

# 等號（=）只會執行一次,要多次要寫在迴圈裡

# 設定幾邊形
n = 6
# 設定往前走多少
m = 100
# 設定角度
angle = 360 / n

for i in range(n):
    # 拿到位置: turtle.pos()
    #取得起始位置
    x, y = turtle.pos()
    #往前走
    turtle.forward(m)
    #取得第二個位置
    oldx, oldy = turtle.pos()
    #回到起始位置
    turtle.goto(x, y)
    #右轉
    turtle.right(angle)
    #往前走
    turtle.forward(m)
    #前往第二個位置
    turtle.goto(oldx, oldy)
    #前往起始位置
    turtle.goto(x, y)

```

```python
import random
import ColabTurtle.Turtle as turtle
turtle.initializeTurtle(initial_speed=10, initial_window_size=(800,500))

n = 100
angle = 360 / n
m = 100

i = 0
while i < n:
    r = random.randint(0, 255)
    g = random.randint(0, 255)
    b = random.randint(0, 255)
    turtle.color(r, g, b)

    x, y = turtle.pos()
    turtle.forward(m)
    x1, y1 = turtle.pos()
    turtle.goto(x, y)
    turtle.right(angle)
    turtle.forward(m)
    turtle.goto(x1, y1)
    turtle.goto(x, y)
    i = i + 1
```
