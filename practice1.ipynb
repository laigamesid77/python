```python
import random

game_list = ["剪刀", "石頭", "布"]

user = int(input("請輸入[0-2]任一數字:"))
com = random.randint(0,2)
#測試輸出
print("我出:"+game_list[user])
print("電腦出:"+game_list[com])

if user == (com + 1) % len(game_list):
    print("我贏")
elif com == (user +1) % len(game_list):
    print("我輸")
else:
    print("平手")
```
