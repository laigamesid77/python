```python
lcom, mcom, rcom = [" ", "o", " "]
n = 10
i  = 0
while i < n:
    # 7是層數,所以層數4要怎麼做才會變成7 ( 7 - i * 2 )
    lcount = i
    mcount = (n * 2 - 1 ) - i * 2
    rcount = i
    total = lcom * lcount + mcom * mcount + rcom * rcount
    print(total)
    i = i + 1
```

```python
#清單內容循環
lcom_list = ["_"]
mcom_list = ["x", "o", "!"]
rcom_list = ["_"]

n = 10
i  = 0
while i < n:
    #一次取一個清單中的值 % 清單的長度
    lcom = lcom_list[i % len(lcom_list)]
    mcom = mcom_list[i]
    rcom = rcom_list[i]

    # 7是層數,所以層數4要怎麼做才會變成7 ( 7 - i * 2 )
    lcount = i
    mcount = (n * 2 - 1 ) - i * 2
    rcount = i

    total = lcom * lcount + mcom * mcount + rcom * rcount
    print(total)
    i = i + 1
```
